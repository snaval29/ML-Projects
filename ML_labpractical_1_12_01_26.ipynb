```python
# Load Iris dataframe
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# create dataframe
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map((dict(enumerate(iris.target_names))))
```

```python
#Display first 5 rows
print("First 5 rows of the DataFrame:")
print(df.head())
print(df.tail())
```

```python
#dataset properties
print("\n Dataset Shape :", df.shape)
print("\n Dataset Info :", df.info())
print("\n Statistical Summary :", df.describe())
print("\n Class distribution :", df['species'].value_counts())
```

```python
#missing Value Check
print("\n Missing Values :", df.isnull().sum())
```

```python
# Data Visualization
# Histogram for each Features

df.iloc[:,:-1].hist(figsize=(10,6))
plt.suptitle("Feature Distribtuion(Histogram)")
plt.tight_layout()
plt.show()
```

```python
plt.figure()
for species in df['species'].unique():
   subset = df[df['species'] == species]
   plt.scatter(subset['sepal length (cm)'],
               subset['sepal width (cm)'],
               label=species)

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.legend()
plt.show()
```

```python
#Boxplot for feature comparison
df.boxplot(column= iris.feature_names, figsize=(10,6))
plt.title("boxplot of Iris Features")
plt.xticks(rotation = 45)
plt.show()
```
